```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('noise(1).jpeg', cv2.IMREAD_GRAYSCALE)

kernel = np.array(
    [[1/4, 1/4],
     [1/4, 1/4]])

convolved_image = cv2.filter2D(image, -1, kernel)
flipped_kernel = np.flip(kernel, -1)
correlated_image = cv2.filter2D(image, -1, flipped_kernel)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.title('Original Image'), plt.imshow(image, cmap='gray'), plt.axis('off')
plt.subplot(1, 2, 2), plt.title('Convolved Image'), plt.imshow(convolved_image, cmap='gray'), plt.axis('off')
plt.show()

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.title('Original Image'), plt.imshow(image, cmap='gray'), plt.axis('off')
plt.subplot(1, 2, 2), plt.title('Correlated Image'), plt.imshow(correlated_image, cmap='gray'), plt.axis('off')
plt.show()
```
